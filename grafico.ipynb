```python
import json
PATH ='classic/results/MountainCar-v0-10000-100-5-100-99-True'
# Open the JSON file in read mode
with open(f'{PATH}/output.json', 'r') as file:
    # Load JSO}N data from the file
    data = json.load(file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Plot the data in a 2x2 grid using Matplotlib
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axs = axs.flatten()

for i,column in enumerate(data.keys()):
    serie = pd.Series(data[column])
    if len(serie) >= 1000:
        serie = serie.rolling(round(len(serie)*0.01),center=False).mean().interpolate()
    
    axs[i].plot(serie, label='Original Data')
    axs[i].set_title(f'{column}')
    axs[i].legend()

plt.tight_layout()
plt.show()
```
